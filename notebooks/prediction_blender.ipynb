```python
import pandas as pd
```

```python
dnn_data = pd.read_csv('../submissions/tfidf_dnn_submission.csv.gz')
lr_data = pd.read_csv('../submissions/logistic_regression_tfidf_submission.csv.gz')
gru_data = pd.read_csv('../submissions/more_features_pooled_gru_fasttext_submission.csv.gz')
```

```python
dnn_data.head()
```

```python
new_df = dnn_data.copy()
new_df.loc[:, ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]] = 0
new_df.head()
```

```python
_classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]

for _class in _classes:
    new_df.loc[:, _class] = (dnn_data.loc[:, _class] + lr_data.loc[:, _class] + gru_data.loc[:, _class] )/3
```

```python
new_df.head()
```

```python
new_df.to_csv('../submissions/3stack_submission.csv.gz', index=False, compression="gzip")
```
